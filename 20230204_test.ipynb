### 実験1

```python
# 評価関数の準備
ac_ei = mb.EI()
ac_lcb = mb.LCB()
ac_random = mb.random_value()
```

```python
# 予測モデルの準備
rf = mb.RandomForestRegressor_bo()
gp = mb.GPRegression_bo()
bwo = mb.BwOForestRegressor_bo()
dummy_model = mb.dummy_model()
```

```python
iterations = 300
n_ensemble = 3
```

```python
model_dic = {'GP':gp,'RandomForest':rf,'bwo':bwo,'random':dummy_model}
ac_dic = { 'LCB':ac_lcb}
eval_list = ['random']
dataset_list = [crossbarrel]
```

```python
# 実験セット
for model_name,model in model_dic.items():
    for dataset in dataset_list:
        for ac_name, ac in ac_dic.items():
            for eval in eval_list:
                bm_opt = mb.benchmarking_mi_opt()
                result_save_name = f'{model_name}_{ac_name}_{eval}_{dataset.dataset_name}_{n_ensemble}ensemble'     
                print(result_save_name)
                bm_opt.calc_pool_opt(model, dataset, ac, sample_select_type= eval,n_ensemble =n_ensemble, result_save_name =result_save_name,iterations = iterations )
    
```

```python
exp_name_list1 = ['GP_LCB_random_CrossedBarrel_3ensemble'
                ,'bwo_LCB_random_CrossedBarrel_3ensemble'
                ,'RandomForest_LCB_random_CrossedBarrel_3ensemble'
                ,'random_LCB_random_CrossedBarrel_3ensemble'
                ] 

perfomance_dataset_dic = {}
for exp_name in exp_name_list1:
    with open(exp_name, 'rb') as f:
        result = pickle.load(f)
        print(f'計算時間 {exp_name}：',result.total_time)
        performance_result = result.calc_avg_performance()
        perfomance_dataset_dic[exp_name] = performance_result

mb.plot_best_trend(perfomance_dataset_dic,crossbarrel,plot_pred = False)
```
